```python
delta = 0.025
scale = 5
xrange = np.arange(-scale, scale, delta)
yrange = np.arange(-scale, scale, delta)
X, Y = np.meshgrid(xrange,yrange)

#
Z=np.sqrt(X**2+Y**2)
fig, ax = plt.subplots()
plt.gca().set_aspect('equal', adjustable='box')
CS = ax.contour(X, Y, Z, range(1,6),cmap='winter')
ax.clabel(CS, inline=True)
#CS = ax.contourf(X, Y, Z)
plt.show()
```

# $\sqrt{x}+\sqrt{y}=\sqrt{a}$

```python
delta = 0.025
xrange = np.arange(-0, 5, delta)
yrange = np.arange(-0, 5, delta)
X, Y = np.meshgrid(xrange,yrange)

Z=(np.sqrt(X)+np.sqrt(Y))**2
fig, ax = plt.subplots()
CS = ax.contour(X, Y, Z, [1,2,3,4], cmap='winter')
ax.clabel(CS, inline=True)
#CS = ax.contourf(X, Y, Z, cmap='summer')
plt.gca().set_aspect('equal', adjustable='box')
ax.clabel(CS, inline=True)
#plt.contour(X, Y, Z, [0])
plt.axis([-0.5, 4.5, -0.5, 4.5])
plt.grid()

plt.show()
```
